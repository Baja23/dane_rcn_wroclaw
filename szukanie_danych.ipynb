```python
import geopandas as gpd

# Krok 1: Pobranie listy warstw
layers_info = gpd.list_layers('Dane/wroclaw_rcn.gml')

# Krok 2: Iteracja po KOLUMNIE 'name' (zawierajƒÖcej nazwy warstw)
with open('Lista_warstw_i_kolumn.txt', 'w', encoding='utf-8') as f:
    for layer_name in layers_info['name']:
        # Krok 3: Wczytanie warstwy
        gdf = gpd.read_file('Dane/wroclaw_rcn.gml', layer=layer_name)
        
        # Krok 4: Zapis informacji
        f.write(f'Warstwa: {layer_name}\n')
        f.write(f'Kolumny: {list(gdf.columns)}\n')
        f.write(f'Liczba wierszy: {len(gdf)}\n')
        f.write('-' * 50 + '\n')
```

```python
import xml.etree.ElementTree as ET

def find_address_tags(gml_path):
    tree = ET.parse(gml_path)
    root = tree.getroot()
    
    print("üîç Szukam powiƒÖza≈Ñ z adresem...")
    
    # Sprawdzamy pr√≥bkƒô dla Dzia≈Çki, Budynku i Lokalu
    for element in root.iter():
        # Je≈õli to dzia≈Çka, budynek lub lokal
        if any(element.tag.endswith(x) for x in ['RCN_Dzialka', 'RCN_Budynek', 'RCN_Lokal']):
            
            # Szukamy dzieci, kt√≥re majƒÖ w nazwie 'adres'
            for child in element:
                if 'adres' in child.tag.lower():
                    print(f"\nZnaleziono w {element.tag.split('}')[-1]}:")
                    print(f"   Tag: {child.tag}")
                    print(f"   Atrybuty: {child.attrib}")
                    # Je≈õli znajdziemy href, to jest to nasz z≈Çoty graal
                    for k, v in child.attrib.items():
                        if k.endswith('href'):
                            print(f"   LINK DO ADRESU: {v}")
                            return # Wystarczy nam jeden przyk≈Çad
                            
find_address_tags('Dane/wroclaw_rcn.gml')
```
