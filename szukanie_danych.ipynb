```python
import geopandas as gpd

# Krok 1: Pobranie listy warstw
layers_info = gpd.list_layers('Dane/wroclaw_rcn.gml')

# Krok 2: Iteracja po KOLUMNIE 'name' (zawierającej nazwy warstw)
with open('Lista_warstw_i_kolumn.txt', 'w', encoding='utf-8') as f:
    for layer_name in layers_info['name']:
        # Krok 3: Wczytanie warstwy
        gdf = gpd.read_file('Dane/wroclaw_rcn.gml', layer=layer_name)
        
        # Krok 4: Zapis informacji
        f.write(f'Warstwa: {layer_name}\n')
        f.write(f'Kolumny: {list(gdf.columns)}\n')
        f.write(f'Liczba wierszy: {len(gdf)}\n')
        f.write('-' * 50 + '\n')
```

```python
from sqlalchemy import create_engine, text, URL
from dotenv import load_dotenv
import os 

load_dotenv()
db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT')
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PASSWORD')
db_name = os.getenv('DB_NAME')

print(db_user)
connection_url = URL.create(
    "postgresql+psycopg2",
    username=db_user,
    password=db_password,
    host=db_host,
    port=db_port,
    database=db_name,
)

#passing all db info to one string variable
#connection_string = f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
engine = create_engine(connection_url)

with open ('enumerations_tables_creation.sql', 'r') as file:
    sql_script = file.read()
    with engine.connect() as connection:
        connection.execute(text(sql_script))
        connection.commit()
```
